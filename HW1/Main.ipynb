```python
from matrix import Matrix 
import matplotlib.pyplot as plt
```

```python
def inverse(M,n):
        a =  [[0 for i in range(n*2)] for j in range(n)]
        for i in range(n):
            for j in range(n):
                if i==j:
                    a[i][j+n]=1
        # Reading matrix coefficients
        for i in range(n):
            for j in range(n):
                a[i][j] = float(M[i,j])

        # Applying Guass Jordan Elimination
        for i in range(n):
            if a[i][i] == 0.0:
                continue
            for j in range(n):
                if i != j:
                    ratio = a[j][i]/a[i][i]
                    for k in range(2*n):
                        a[j][k] = a[j][k] - ratio * a[i][k]

    # Row operation to make principal diagonal element to 1
        for i in range(n):
            div = a[i][i]
            for j in range(2*n):
                a[i][j] = a[i][j]/div
        ans =  Matrix(dim = (n, n), value = 0)  
        for i in range(n):
            for j in range(n):
                ans[i,j] = a[i][n+j]
        return ans    
           
def identitymatrix(n):
    I =   Matrix(dim = (n, n), value = 0)
    for i in range(n):
        for j in range (n):
            if(i==j):
                I[i,j] = 1
    return I 
        
# def printLine(value):
#     function = ""
#     for i in range(len(value) - 1):
#         function = function + str(value[i, 0]) + "x^" + str(value.row - 1 - i) + " + "
#     function = function + str(value[ value.row - 1, 0])
#     print("Fitting ine : ", value)    

def error(A, line, b):
    p_y = A @ line
    err = p_y - b
    eSum = 0
    for i in range(b.row):
        eSum += err[i, 0] * err[i, 0]
    print("Total error : ", eSum)


def plotline(x, y, line, n):
    p_y = [0 for i in range (len(x))]
    for i in range(len(x)):
        t = 0
        for j in range(n):
            t += line[j, 0] * x[i] ** (n - j - 1)
        p_y[i] = t
    
    plt.grid()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(x, y, color="red")
    plt.plot(x, p_y,color="green") 
    plt.show()
```

```python
# LSE method
inputA = A.transpose() @ A + Lambda * identitymatrix(n) 
line = inverse(inputA,n) @ A.transpose()@ b
print(line)
plotline(x,y,line,n)
error(A, line, b)
```
